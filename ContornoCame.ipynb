```python
from sympy import *
from sympy.plotting import plot_parametric
%matplotlib inline

θ = symbols('theta')

def GetElev(curva, h, β, revert = 0):
    if (curva == 'harmonica'):
        u = 0.5*h*( 1 - cos(pi*θ/β) )
    elif (curva == 'cicloide'):
        u = h*( θ/β - (1/(2*pi))*sin(2*pi*θ/β) )
    elif (curva == 'duplaharmonica'):
        u = 0.5*h*( 1 - cos(pi*θ/β) - 0.25*(1 - cos(2*pi*θ/β)) )
    elif (curva[0] in ['3','4']):
        curva = curva.replace('-',"")
        k = [int(i) for i in curva]
        if (len(k) == 2):
            a,b = k[0],k[1]
            u = h*(b*(θ/β)**a - a*(θ/β)**b)
        elif (len(k) == 3):
            a,b,c = k[0],k[1],k[2]
            u = 0.5*h*(b*c*(θ/β)**a - 2*a*c*(θ/β)**b + a*b*(θ/β)**c)
        elif (len(k) == 4):
            a,b,c,d = k[0],k[1],k[2],k[3]
            u = 0.5*h*( (1/3)*b*c*d*(θ/β)**a - a*c*d*(θ/β)**b + a*b*d*(θ/β)**c - (1/3)*a*b*c*(θ/β)**d )
    else:
        print('ERRO --> função inexistente')
        return 0
    
    if (revert == 1):
        u = u.subs(θ,β-θ)
    
    return u

def PlotPerfil(Elev, Ret, βₑ, βᵣₛ, βᵣ, φ, h, Rₚ, Rᵣ):
    f = GetElev(Elev, h, βₑ)
    g = GetElev(Ret, h, βᵣ); g = g.subs(θ,βₑ+βᵣ-θ)
    fₜ = diff(f,θ); φf = atan( fₜ/(f+Rₚ) )
    gₜ = diff(g,θ); φg = atan( gₜ/(g+Rₚ) )
    
    xc = -(Rₚ + f)*sin(βₑ - θ)# - Rᵣ*sin(φf + βₑ - θ), houve espelhamento em y --> { -(Rₚ...}
    yc = (Rₚ + f)*cos(βₑ - θ)# - Rᵣ*cos(φf + βₑ - θ)
    p1 = plot_parametric(xc, yc, (θ,0,βₑ), show = False)
    
    xc = -(Rₚ + g)*sin(βₑ - θ)# - Rᵣ*sin(φg + βᵣ - θ), houve espelhamento em y --> { -(Rₚ...}
    yc = (Rₚ + g)*cos(βₑ - θ)# - Rᵣ*cos(φg + βᵣ - θ)
    p2 = plot_parametric(xc, yc, (θ,βₑ,βₑ+βᵣ), show = False)
    
    if (pi/2-βₑ > 0):
        p3 = plot_parametric(-Rₚ*cos(θ), Rₚ*sin(θ), (θ,0,pi/2-βₑ), show= False)
        p3[0].line_color='green'
        p1.extend(p3)

    p4 = plot_parametric(-Rₚ*cos(θ), Rₚ*sin(θ), (θ,pi/2+βᵣ,2*pi + pi/2-βₑ), show= False)
    p5= plot_parametric(0.1*Rp*cos(θ), 0.1*Rp*sin(θ), (θ,0,2*pi), show= False)
    
    #p1.axis = False
    p1.aspect_ratio=(1,1)
    
    xc = -(Rₚ + f)*sin(βₑ - θ) + Rᵣ*sin(φf + βₑ - θ)
    yc = (Rₚ + f)*cos(βₑ - θ) - Rᵣ*cos(φf + βₑ - θ)
    p6 = plot_parametric(xc, yc, (θ,0,βₑ), show = False)
    
    xc = -(Rₚ + g)*sin(βₑ - θ) + Rᵣ*sin(φg + βₑ - θ)
    yc = (Rₚ + g)*cos(βₑ - θ) - Rᵣ*cos(φg + βₑ - θ)
    p7 = plot_parametric(xc, yc, (θ,βₑ,βₑ+βᵣ), show = False)
    
    if (pi/2-βₑ > 0):
        p8 = plot_parametric(-(Rₚ-Rᵣ)*cos(θ), (Rₚ-Rᵣ)*sin(θ), (θ,0,pi/2-βₑ), show= False)
        p8[0].line_color='red'
        p1.extend(p8)
    p9 = plot_parametric(-(Rₚ-Rᵣ)*cos(θ), (Rₚ-Rᵣ)*sin(θ), (θ,pi/2+βᵣ,2*pi + pi/2-βₑ), show= False)
    
    p1[0].line_color='green'
    p2[0].line_color='green'
    p4[0].line_color='green'
    p6[0].line_color='red'
    p7[0].line_color='red'
    p9[0].line_color='red'
    
    print( 'Elevação por '+Elev+' e Retorno por '+Ret )
    print( 'βₑ = '+str(180*βₑ/pi)+'°'+', βᵣ = '+str(180*βᵣ/pi)+'°'+' e βᵣᵢ = '+str(180*(2*pi-βₑ-βᵣ)/pi)+'°' )
    
    p1.extend(p2)
    p1.extend(p4)
    p1.extend(p5)
    p1.extend(p6)
    p1.extend(p7)
    p1.extend(p9)
    p1.show()
```

```python
Elev = 'harmonica'
Ret = 'cicloide'
βₑ = 2*pi/3; βᵣₛ = 0; βᵣ = 2*pi/3
φ = pi/6
h = 8; Rₚ = 18; Rᵣ = 5

PlotPerfil(Elev, Ret, βₑ, βᵣₛ, βᵣ, φ, h, Rₚ, Rᵣ)
```
